# Get an text file from a Youtube video

The code below is extracting an text file from a youtube video and adding it in the input data folder

```python
# Install the youtube-transcript-api package using pip
pip install youtube-transcript-api
```

### 1. Setup — Imports & Data Folder

```python
import re
from pathlib import Path
from dataclasses import dataclass
from youtube_transcript_api import YouTubeTranscriptApi

DATA_DIR = Path("input data")
DATA_DIR.mkdir(exist_ok=True)
```

### 2. Configuration — Data Structure


```python
@dataclass
class SourceDocument:
    source_type: str
    origin: str
    raw_text: str
```

### 3. Core Functions — Slugify & Save


```python
def _slugify(url: str) -> str:
    """Turns a URL into a safe filename, e.g. www-youtube-com-watch-v-AWukNG1UpEs.txt"""
    slug = re.sub(r"^https?://", "", url)
    slug = re.sub(r"[^a-zA-Z0-9]+", "-", slug).strip("-").lower()
    return slug[:100]


def save_to_file(doc: SourceDocument, folder: Path = DATA_DIR) -> Path:
    """Saves a SourceDocument's raw_text to 'input data/<slug>.txt' and returns the path."""
    filename = f"{_slugify(doc.origin)}.txt"
    out_path = folder / filename
    out_path.write_text(doc.raw_text, encoding="utf-8")
    return out_path
```

### 4. Run — Fetch Transcript & Save


```python
video_url = "https://www.youtube.com/watch?v=AWukNG1UpEs"
video_id = "AWukNG1UpEs"

ytt_api = YouTubeTranscriptApi()
fetched_transcript = ytt_api.fetch(video_id)
text = " ".join(snippet.text for snippet in fetched_transcript)

doc = SourceDocument(
    source_type="youtube",
    origin=video_url,
    raw_text=text
)
saved_path = save_to_file(doc)

print(f"Extracted {len(text.split())} words")
print(f"Saved to: {saved_path}")
```
